# Previsão de dias populição da estação popuente de Santana - SP

```python
dados1 = pd.Series(np.random.randn(1000), index=pd.date_range('2/4/2020', periods=1000))
dados1 = dados1.cumsum()
plt.title("Previsão de dias da poluição da estação poluente de Santana 2020")
plt.xlabel("Anos")
plt.ylabel("Dias e anos")
dados1.plot() 
```
