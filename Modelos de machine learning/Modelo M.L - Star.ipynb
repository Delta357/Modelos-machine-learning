```python
y_dados = dados["target_class"].values
x_dados = dados.drop(["target_class"],axis=1)
x1 = (x_dados - np.min(x_dados))/(np.max(x_dados)-np.min(x_dados))
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_dados, y_dados, test_size = 0.3,random_state=1)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_train
```

**Naive bayes**

```python
from sklearn.naive_bayes import GaussianNB

naive_bayes = GaussianNB()
naive_bayes.fit(x_train, y_train)
naive_bayes_p = naive_bayes.predict(x_test)
print(naive_bayes_p)
```

**K-NN**

```python
from sklearn.neighbors import KNeighborsClassifier

KN = KNeighborsClassifier(n_neighbors = 13)
KN.fit(x_train, y_train)
KN_P = KN.predict(x_test)
print(KN_P)
```

**Random forest**

```python
from sklearn.ensemble import RandomForestClassifier

arvore1 = RandomForestClassifier(n_estimators = 100)
arvore1.fit(x_train, y_train)
arvore1_predicao = arvore1.predict(x_test)
print(arvore1_predicao)
```

```python
arvore2 = RandomForestClassifier(n_estimators = 100, random_state = 10)
arvore2.fit(x_train, y_train)
arvore2_predicao2 = arvore2.predict(x_test)
print(arvore2_predicao2)
```

```python
from sklearn.metrics import confusion_matrix,classification_report

matrix = classification_report(y_test, naive_bayes_p)
print("Classificaçao do Naive Bayes: \n", matrix)
```

```python
matrix = classification_report(y_test, KN_P)
print("Classificaçao do KNN \n", matrix)
```

```python
matrix = classification_report(y_test, arvore1_predicao)
print("Classificação do Random Forest \n", matrix)
```
