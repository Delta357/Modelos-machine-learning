```python
def p1a(titulo, labelx, labely, x, y, dataset):
    sns.set_palette("Accent")
    sns.set_style('darkgrid')

    df = sns.lineplot(x=x, y=y, data=dataset)
    
    df.figure.set_size_inches(20, 10)
    df.set_title(titulo, loc="left", fontsize=20)
    df.set_xlabel(labelx, fontsize=20)
    df.set_ylabel(labely, fontsize=20)
    df = df
```

```python
p1a("aÃ§oes da petrobras", "AÃ§Ãµes 2020", "Alta", 'Baixa', 'Aberta', dados)
```
